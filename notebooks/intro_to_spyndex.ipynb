```python
# Open a dataset (in this case a pandas.DataFrame)
df = spyndex.datasets.open("spectral")
df.head()
```

```python
df['class'].value_counts()
```

```python
# Compute the desired spectral indices
idx = spyndex.computeIndex(
    index = ["NDVI","NDWI","NDBI"],
    params = {
        "N": df["SR_B5"],
        "R": df["SR_B4"],
        "G": df["SR_B3"],
        "S1": df["SR_B6"]
    }
)
```
